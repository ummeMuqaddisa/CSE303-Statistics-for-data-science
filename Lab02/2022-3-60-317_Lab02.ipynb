```python
vowels="aeiouAEIOU"
zero_vowels=""
for char in string:
  if char not in vowels:
    zero_vowels=zero_vowels+char

print(zero_vowels)
```

```python
words=string.split()
new_words=[]

for i in words:
  if len(i)<5:
    new_words.append(i)
print(new_words)
```

```python
words=string.split()
word_len={}

for i in words:
  word_len[i]=len(i)
print(word_len)
```

```python
numbers=[]

for i in range (1,1001):
  for j in range(2,9):
    if i%j==0:
      numbers.append(i)
      break


print(numbers)
```

```python
highest_divisor={}
for i in range(1,1001):
  max_div=0
  for j in range (2,10):
    if i%j==0 and j>max_div:
      max_div=j
  highest_divisor[i]=max_div
print(highest_divisor)
```
